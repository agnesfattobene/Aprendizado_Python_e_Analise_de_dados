# 1. Formatação de Print

```python
paciente = 'Douglas'
idade = 47
peso = 150
altura = 1.85
imc = peso / (altura ** 2)
```

```python
print(imc)
```

```python
# print com f-string
print(f'O nome do "paciente" é {paciente}') # esse tipo de formatação de print serve para combinar textos gerais com os dados de variáveis no mesmo print
```

```python
'''
Exemplo imprimir o texto: "O nome do paciente é ______,
ele tem ______ anos de idade, seu peso é _____ kg, sua
altura é ______ em m e seu imc é _______"
'''
print(f'O nome do paciente é {paciente}, ele tem {idade} anos de idade, seu peso é {peso} kg, sua altura é {altura} em m e seu imc é {imc:.2f}')
```

```python
'''
Exercício: imprima a mensagem anterior com o seguinte adicional
se o valor do imc for menor ou igual a 25 adicione o texto
"peso adequado para suas características", caso contrário
"o seu peso pode não estar adequado para suas características"
'''
# Gabarito
if imc <= 25:
  print(f'O nome do paciente é {paciente}, ele tem {idade} anos de idade, seu peso é {peso} kg, sua altura é {altura} em m e seu imc é {imc:.2f} e seu peso é adequado para suas características')
else:
  print(f'O nome do paciente é {paciente}, ele tem {idade} anos de idade, seu peso é {peso} kg, sua altura é {altura} em m e seu imc é {imc:.2f} e seu peso pode não estar adequado para suas características')
```

## 2.1 Listas

```python
# Uma lista é um objeto que pode ser vazio, ou ter um ou mais itens
lista = ['dado1', 'dado2','dado3','dado4', 'dado_n']
#índices    0         1       2       3       4       ou:
#          -5        -4      -3      -2      -1
print(f'O quinto elemento da lista é {lista[4]}')
print(f'O quinto elemento da lista é {lista[-1]}')
```

```python
# Métodos de listas
# função len (Contagem do número de dados da lista)
print(len(lista))
```

```python
#inserindo dados dentro de uma lista
clientes = [] # cria a lista vazia "clientes"
print(len(clientes)) # imprime a quantidade atual da lista
print(clientes) # imprime os valores atuais da lista
# inserindo o cliente Douglas pelo método append
clientes.append('Douglas') # adiciona o texto "Douglas"
print(len(clientes)) # imprime a quantidade atual da lista
print(clientes) # imprime os valores atuais da lista
# inserindo a cliente Daniela
clientes.append('Daniela') # adiciona o texto "Douglas"# por padrão esse comando colocará o dado na última posição da lista
print(len(clientes)) # imprime a quantidade atual da lista
print(clientes) # imprime os valores atuais da lista
```

```python
'''
Caso de uso: O que deveríamos fazer se quiséssemos inserir
o nome Daniela antes de Douglas
'''
#inserindo dados dentro de uma lista
clientes = [] # cria a lista vazia "clientes"
print(len(clientes)) # imprime a quantidade atual da lista
print(clientes) # imprime os valores atuais da lista
# inserindo o cliente Douglas pelo método append
clientes.append('Douglas') # adiciona o texto "Douglas"
print(len(clientes)) # imprime a quantidade atual da lista
print(clientes) # imprime os valores atuais da lista
# inserindo a cliente Daniela antes de Douglas
clientes.insert(0,'Daniela')
print(len(clientes)) # imprime a quantidade atual da lista
print(clientes) # imprime os valores atuais da lista
```

```python
# Removendo elementos da lista
# Método 1 (pop)
print(clientes)
clientes.pop() # o método pop sem argumento remove por padrão o último dado da lista
print(clientes)
```

```python
clientes = ['Daniela', 'Douglas']
print(clientes)
# excluir o 1º elemento da lista (índice 0)
clientes.pop(0) # excluindo o dado do índice 0
print(clientes)
```

```python
# excluindo um elemento com o método remove
clientes = ['Daniela', 'Douglas']
print(clientes)
clientes.remove('Douglas')
print(clientes)
```

```python
# Excluindo todos os dados de uma lista
clientes = ['Daniela', 'Douglas']
print(clientes)
clientes.clear()
print(clientes)
```

```python
# criando cópia de uma lista
clientes = ['Daniela', 'Douglas']
print(clientes)
bkp = clientes.copy()
print(bkp)
print(clientes)
clientes.clear()
print(clientes)
print(bkp)
```
