## **Dicionario de Dados**


*  Title: Título do vídeo.
*  Video ID: O identificador de vídeo.
*  Published At: A data em que o vídeo foi publicado YYYY-MM-DD.
*  Keyword: a palavra-chave associada ao vídeo.
*  Likes: o número de curtidas que o vídeo recebeu. Se este valor for -1, as curtidas não serão visíveis publicamente.
*  Comments: o número de comentários que o vídeo tem. Se esse valor for -1, o criador do vídeo desativou os comentários.
*  Views: o número de visualizações que o vídeo obteve.

Mais informações: https://www.kaggle.com/datasets/advaypatil/youtube-statistics?select=videos-stats.csv

```python
# Visualizção aleatoria de dados
df.sample(5)
```

```python
# Renomeando colunas para tradução
df.rename(columns={'Unnamed: 0': 'linha',
                   'Title': 'titulo',
                   'Video ID':'video_id',
                   'Published At': 'data_publicacao',
                   'Keyword':'palavra_chave',
                   'Likes':'qtd_curtidas',
                   'Comments': 'qtd_comentarios',
                   'Views': 'qtd_visualizacoes'}, inplace=True)
```

```python
#checagem
df.info()
```

```python
# Checando dados nulos
df.isna().sum()
```

```python
# Checando dados nulos
df.isnull().sum()
```

```python
# Calcular a porcentagem de valores nulos em cada coluna
porcentagem_nulos = (df.isnull().sum() / len(df)) * 100

# Exibir a porcentagem de valores nulos em cada coluna
print(round(porcentagem_nulos,1))
```

```python
# Removendo linhas com valores NaN
df = df.dropna()
```

```python
# Checando se de fato os valores nulos foram eliminados
df.isnull().sum()
```

```python
df.sample()
```

```python
df.info()
```

```python
# Verificando se os dados são unicos na coluna 'video_id'
# Pelo fato de ter aparecido 'false', significa que temos dados duplicados
df.video_id.is_unique
```

```python
# Encontrando as linhas duplicadas com base na coluna 'video_id'
df[df.duplicated(subset='video_id', keep=False)]
```

```python
# verificando tamanho do dataframe
df.shape
```

```python
# eliminando duplicadas via video_id
df = df.drop_duplicates(subset='video_id')
```

```python
# verificando
df.shape
```

```python
# Verificando
print('A coluna video_id possui valores unicos? Resposta: ', df.video_id.is_unique)
```

```python
# Verificando valores unicos na coluna 'titulo
# Como a resposta foi false, significa que temos titulos iguais na coluna
print('A coluna titulo possui valores unicos? Resposta: ', df.titulo.is_unique)
```

```python
# Encontrar linhas duplicadas com base na coluna 'titulo'
df[df.duplicated(subset='titulo', keep=False)]
```

```python
# Eliminando duplicidade do titulo (cabe verificação mais aprodundada)
df = df.drop_duplicates(subset='titulo')
```

```python
# Verificando tamanho
df.shape
```

```python
# Verificando se dados são unicos na coluna na coluna 'titulo'
print('Tem apenas valores únicos na coluna titulo? Resposta', df.titulo.is_unique)
```

```python
df.sample(5)
```

```python
# Eliminação de colunas desnecessárias primeiramente
# O parâmetro axis=1 indica que as colunas devem ser removidas
# (em oposição a linhas, que seria axis=0).
df.drop(['linha',
         'video_id'], axis=1, inplace=True)
```

```python
# Outra forma de eliminar colunas
#df = df.drop(['linha',
#         'video_id'], axis=1)
```

```python
# Verificação
df.head(1)
```

```python
# Verificação dos tipos de dados
df.dtypes
```

```python
# Convertendo valores numéricos de float para int
df['qtd_curtidas'] = df['qtd_curtidas'].astype(int)
df['qtd_comentarios'] = df['qtd_comentarios'].astype(int)
df['qtd_visualizacoes'] = df['qtd_visualizacoes'].astype(int)
```

```python
# Convertendo a coluna de data_publicação para datetime
df['data_publicacao'] = pd.to_datetime(df['data_publicacao'])
```

```python
# Checagem de conversão
df.dtypes
```

```python
# Separando dia, mes e ano em colunas diferentes
df['ano'] = df.data_publicacao.dt.year

df['mes'] = df.data_publicacao.dt.month

df['dia'] = df.data_publicacao.dt.day
```

```python
df.head()
```

```python
df.dtypes
```

```python
# Converter as colunas para int - str
df['ano'] = df['ano'].astype(str)
df['mes'] = df['mes'].astype(str)
df['dia'] = df['dia'].astype(str)
```
