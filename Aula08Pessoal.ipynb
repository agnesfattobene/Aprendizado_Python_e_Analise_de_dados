```python
 import math

def calcular_media(notas):
    return sum(notas) / len(notas)

def calcular_desvio_padrao(notas, media):
    soma_diferencas_quadradas = sum((nota - media) ** 2 for nota in notas)
    variancia = soma_diferencas_quadradas / len(notas)
    desvio_padrao = math.sqrt(variancia)
    return desvio_padrao

def main():
    try:
        n = int(input("Informe a quantidade de notas: "))
        if n <= 0:
            print("Por favor, informe um número válido de notas.")
            return

        notas = []
        for i in range(n):
            nota = float(input(f"Informe a nota {i + 1} (0 - 10): "))
            if 0 <= nota <= 10:
                notas.append(nota)
            else:
                print("Nota inválida. Por favor, informe uma nota entre 0 e 10.")
                return

        media = calcular_media(notas)
        desvio_padrao = calcular_desvio_padrao(notas, media)

        print(f"\nMédia: {media:.2f}")
        print(f"Desvio Padrão: {desvio_padrao:.2f}")

    except ValueError:
        print("Por favor, informe valores numéricos válidos.")

if __name__ == "__main__":
    main()
```

```python
import math

def calcular_media(notas):
    return sum(notas) / len(notas)

def calcular_desvio_padrao(notas, media):
    variancia = sum((nota - media) ** 2 for nota in notas) / len(notas)
    desvio_padrao = math.sqrt(variancia)
    return desvio_padrao

n = int(input("Informe a quantidade de notas: "))

notas = [float(input(f"Informe a nota {i + 1} (0 - 10): ")) for i in range(n) if 0 <= nota <= 10]

media = calcular_media(notas)
desvio_padrao = calcular_desvio_padrao(notas, media)

print(f"\nMédia: {media:.2f}")
print(f"Desvio Padrão: {desvio_padrao:.2f}")
```

```python
n = int(input("Informe a quantidade de notas: "))

notas = []
for i in range(n):
    nota = float(input(f"Informe a nota {i + 1} (0 - 10): "))
    while nota < 0 or nota > 10:
        print("Nota inválida. Por favor, informe uma nota entre 0 e 10.")
        nota = float(input(f"Informe a nota {i + 1} (0 - 10): "))
    notas.append(nota)

media = sum(notas) / n
desvio_padrao = math.sqrt(sum((nota - media) ** 2 for nota in notas) / n)

print(f"\nMédia: {media:.2f}")
print(f"Desvio Padrão: {desvio_padrao:.2f}")
```
