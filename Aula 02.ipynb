```python
# Tipos de variáveis
# Variável de texto (str) - String - (Variável qualitativa)
# Variável numérica inteira (int) - (Variável Quantitativa Discreta)
# Variável numérica fracionária (float) - (Variável Quantitativa Contínua)

cliente = "Douglas"  # criando a variável de texto cliente que recebe o nome Douglas
print(cliente) # para mostrar o que tem dentro da variável precisamos usar o nome da variável

peso = 79 # Criando a variável peso com o dado numérico inteiro 79
print(peso) # mostrando o conteúdo da variável peso

altura = 1.85 # Criando a variável altura que recebe o valor 1.85
print(altura) # exibindo o valor da variável altura
```

```python
# Operadores matemáticos
# soma (+)
# subtração (-)
# multiplicação (*)
# divisão (/)
# parte inteira da divisão (//) (para descartar o valor quebrado)
# resto da divisão (%)
# potenciação (**)

# exemplo crie uma variável chamada soma que receba a conta 3 + 2
soma = 3 + 2
print(soma)
# exemplo 2 crie uma variável chamada sub que receba a conta 3 - 2
sub = 3 - 2
print(sub)
# exemplo 3 crie uma variável chamada mult que receba o cálculo 3 * 2
mult = 3 * 2
print(mult)
# exemplo 4 crie uma variável chamada div que receba o cálculo 3 / 2
div = 3 / 2
print(div)
# exemplo 5 crie uma variável chamada divint que receba o cálculo 3 // 2
divint = 3 // 2
print(divint)
# exemplo 6 crie uma variável chamada resto que receba o cálculo 3 % 2
resto = 3 % 2
print(resto)
# exemplo 7 crie uma variável chamada pot que receba o cálculo 3 ** 2
pot =  3 ** 2
print(pot)
```

```python
# Bônus radiciação

# crie uma variável raiz2 que receba o cálculo da raiz quadrada de 4

raiz2 = 4 ** (1/2)
print(raiz2)
```

```python
# Operadores de atribuição
# operador de atribuição direta (=)
cliente = 'Douglas' # atribuindo o valor Douglas para a variável cliente
cliente = 'Daniela' #atribuindo o valor Daniela para a variável cliente (substitui o valor antigo)
print(cliente)
```

```python
# operadores de atribuição continuação
idade = 47 # Atribuição o valor 47 para a variável idade
print(idade) # mostrando o valor atual da variável idade
```

```python
print(idade) # mostra o valor atual da variável idade (se ela já existir)
# Atualize o valor da variável idade em um ano (some 1)
idade = idade + 1 # acrescenta mais um ao valor atual da variável idade
print(idade)
# alternativa: operador de atribuição de soma (+=)
idade += 1 # acrescenta mais um ao valor atual da variável idade
print(idade)
```

```python
# subtraia 1 do valor da variável idade
idade = 47 # torna o estado atual = 47
print(idade) # imprime o valor do estado atual (47)
idade = idade - 1 # diminui em 1 o estado atual vai para (46)
print(idade) # imprime o novo valor do estado atual (46)
idade = 47 # muda o valor do estado atual da variável para 47
idade -= 1 # muda o valor do estado atual para 46
print(idade) # imprime o novo valor do estado atual (46)
```

```python
# Operadores de comparação
# operador de igualdade (==)
'''
Exemplo: crie uma variável chamada cliente e atribua a ela o valor Douglas,
depois crie uma variável chamada fornecedor e atribua a ela o valor Douglas,
após isso crie uma variável chamada comp e compare se fornecedor é ou não
igual a cliente
'''
cliente = 'Douglas' # cria a variável cliente e atribui o valor Douglas
fornecedor = 'Douglas' # cria a variável fornecedor e atribui o valor Douglas
comp = cliente == fornecedor # cria a variável comp e testa se o valor de cliente é igual ao valor de fornecedor
print(comp) # imprime o resultado da comparação (verdadeiro "True" ou Falso "False")
```

```python
# Operadores de comparação
# operador de diferença (!=)
'''
Exemplo: crie uma variável chamada cliente e atribua a ela o valor Douglas,
depois crie uma variável chamada fornecedor e atribua a ela o valor Douglas,
após isso crie uma variável chamada comp2 e compare se fornecedor é ou não
diferente de cliente
'''
cliente = 'Douglas' # cria a variável cliente e atribui o valor Douglas
fornecedor = 'Douglas' # cria a variável fornecedor e atribui o valor Douglas
comp2 = cliente != fornecedor # cria a variável comp2 e testa se o valor de cliente é diferente ao valor de fornecedor
print(comp2) # imprime o resultado da comparação nesse caso uma comparação de diferença (verdadeiro "True" ou Falso "False")
```

```python
# Comparação de maior que (>)
idade1 = 47 # cria a variável idade1 e lhe atribui o valor 47
idade2 = 44 # cria a variável idade2 e lhe atribui o valor 44
comp3 = idade1 > idade2 #compara se idade1 é maior que idade 2 e armazena na variável comp3
print(comp3) # imprime o resultado de comp3
comp4 = idade2 > idade1#compara se idade2 é maior que idade 1 e armazena na variável comp4
print(comp4) # imprime a comparação comp4
```

```python
# Comparação de menor que (<)
idade1 = 47 # cria a variável idade1 e lhe atribui o valor 47
idade2 = 44 # cria a variável idade2 e lhe atribui o valor 44
comp5 = idade1 < idade2 #compara se idade1 é menor que idade 2 e armazena na variável comp5
print(comp5) # imprime o resultado de comp5
comp6 = idade2 < idade1 #compara se idade2 é menor que idade 1 e armazena na variável comp6
print(comp6) # imprime a comparação comp6
```

```python
# Comparação de maior ou igual que (>=)
idade1 = 47 # cria a variável idade1 e lhe atribui o valor 47
idade2 = 47 # cria a variável idade2 e lhe atribui o valor 44
comp7 = idade1 >= idade2 #compara se idade1 é maior ou igual que idade 2 e armazena na variável comp7
print(comp7) # imprime o resultado de comp7
comp8 = idade2 >= idade1 #compara se idade2 é maior ou igaul que idade 1 e armazena na variável comp8
print(comp8) # imprime a comparação comp8
```

```python
# Comparação de menor ou igual que (<=)
idade1 = 47 # cria a variável idade1 e lhe atribui o valor 47
idade2 = 47 # cria a variável idade2 e lhe atribui o valor 44
comp9 = idade1 <= idade2 #compara se idade1 é menor que idade 2 e armazena na variável comp9
print(comp9) # imprime o resultado de comp9
comp10 = idade2 <= idade1 #compara se idade2 é menor que idade 1 e armazena na variável comp10
print(comp10) # imprime a comparação comp10
```

```python
# Operadores lógicos
# Operador E (and)
comp11 = 3 > 2 and 2 == 3
#         V           F   traz com resultado False V and F = F
print(comp11)
# V and F = F
# V and V = V
# F and V = F
# F and F = F
comp12 = 3 > 2 and 3 == 3
#         V           V   traz com resultado True V and V = V
print(comp12)
comp13 = 3 < 2 and 3 == 3
#         F           V   traz com resultado False F and V = F
print(comp13)
comp14 = 3 < 2 and 3 != 3
#          F         F
print(comp14)
```
