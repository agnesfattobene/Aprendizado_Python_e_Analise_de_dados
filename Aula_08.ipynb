```python
# Fundamentos de Python - Aprofundamento de técnicas
'''
Crie um programa que receba do usuário cinco valores de notas (0 - 10)
calcule a média e informe o usuário a média aritmética dos valores informados
'''
n1 = float(input('Informe a 1ª nota de 0 a 10: '))
n2 = float(input('Informe a 2ª nota de 0 a 10: '))
n3 = float(input('Informe a 3ª nota de 0 a 10: '))
n4 = float(input('Informe a 4ª nota de 0 a 10: '))
n5 = float(input('Informe a 5ª nota de 0 a 10: '))
media = (n1+n2+n3+n4+n5)/5
print(f'A média aritmética dos valores informados é {media:.2f}')
```

```python
'''
Crie um programa que receba do usuário n valores de notas (0 - 10)
calcule a média e informe o usuário a média aritmética dos valores informados
Método de resolução com while
'''
notas = [] # Cria a lista vazia que receberá as notas
c = 0 # Cria a variável contadora que deverá ser atualizada a cada novo input do usuários
qtde = int(input('Informe a quantidade de notas que deseja inserir no sistema: ')) # cria a variável que o usuário informará a qtde de notas a ser inserida no sistema
while c < qtde: # cria a condição para o laço ser executado
  notas.append(float(input(f'Informe a {c+1}ª nota: '))) # cria o input com cada nota e as insere na lista notas criada anteriormentes
  c += 1 # Atualiza a variável contadora a cada novo input
media = sum(notas)/len(notas) # Calcula a média aritmética das notas
print(f'A média aritmética dos valores informados é {media:.2f}') # Imprime a média
```

```python
'''
Crie um programa que receba do usuário n valores de notas (0 - 10)
calcule a média e informe o usuário a média aritmética dos valores informados
Método de resolução sem while
'''

dados = input('Informe as notas separadas por espaços: ') # recebe todas as notas em um único input
print(dados) # print de controle para ver como estão os dados
print(type(dados)) # print de controle para ver o tipo dos dados
notas = dados.split() # picotando os dados usando os espaços com separadores e os inserindo em uma lista chamada notas
print(notas)  # print de controle para ver como estão os dados
print(type(notas)) # print de controle para ver o tipo dos dados
print(type(notas[0]))# print de controle para ver o tipo dos dados
for i in range(len(notas)): # criando um índice (i) e um intervalo com o tamanho da lista para ser percorrido
  notas[i] = float(notas[i]) # transformando cada valor encontrado na lista em float
media = sum(notas)/len(notas) # Calcula a média aritmética das notas
print(f'A média aritmética dos valores informados é {media:.2f}') # Imprime a média
```

```python
'''
Crie um programa que receba do usuário n valores de notas (0 - 10)
calcule a média e informe o usuário a média aritmética, e o desvio padrão
dos valores informados
'''
dados = input('Informe as notas separadas por espaços: ') # recebe todas as notas em um único input
print(dados) # print de controle para ver como estão os dados
print(type(dados)) # print de controle para ver o tipo dos dados
notas = dados.split() # picotando os dados usando os espaços com separadores e os inserindo em uma lista chamada notas
print(notas)  # print de controle para ver como estão os dados
print(type(notas)) # print de controle para ver o tipo dos dados
print(type(notas[0]))# print de controle para ver o tipo dos dados
for i in range(len(notas)): # criando um índice (i) e um intervalo com o tamanho da lista para ser percorrido
  notas[i] = float(notas[i]) # transformando cada valor encontrado na lista em float
media = sum(notas)/len(notas) # Calcula a média aritmética das notas
nq = [] # criando a lista vazia que receberá as diferenças ao quadrado de cada nota para a média
for nota in notas: # para cada nota na lista notas faça alguma coisa
  nq.append((nota - media)**2) # pegando cada nota subtraindo a média, elevando ao quadrado e inserindo na lista (nq) criada anteriormente
desv = (sum(nq)/(len(nq)-1))**(1/2)
print(f'A média aritmética dos valores informados é {media:.2f}') # Imprime a média
print(f'O desvio padrão dos valores informados é {desv:.2f}')
```

```python
'''
Exemplo 02 -

Elabore um programa que que receba um número inteiro do usuário e verifique
se o número informado é par ou ímpar, se for par imprima uma mensagem dizendo
que o número é par e encerre o loop, caso seja ímpar exiba uma mensagem
informando que o número não é par e peça para o usuário digitar outro número,
até que o número digitado pelo usuário seja par.

'''
# While True:
while True: # Cria um loop mais duradouro
  n = int(input('Informe um número inteiro e par: ')) # solicita ao usuário um número, (obs, como esse input está dentro do While True, o programa continuará solicitando um número enquanto a condição exigida não for satisfeita )
  if n % 2 == 0: # cria-se a condição para que o laço seja interrompido
    print(f'{n} é um número par!!! Parabéns!!!') # exibe a mensagem caso a condição seja atendida
    break # interrope o laço se a condição for atendida
  else: # Cria-se o cenário para caso a condição não seja atendida
    print(f'{n} não é um número par e inteiro!!! digite novamente') # exibe a mensagem caso a condição não seja atendida
```

```python
'''
Exemplo:
Elabore um script que receba do usuário um número qualquer e devolva ao usuário
a mensagem "Parabéns você digitou um número!!!"
'''
while True: # cria o laço duradouro
  try: # cria a condição se verdadeiro
    n = float(input('Informe um número qualquer:')) # recebe do usuário uma informação
    print(f'Parabéns você digitou o número {n}') # imprime a informação digitada pelo usuário
    break # interrompe o laço
  except: # cria a condição se falso
    print(f'o valor digitado não é válido!! digite um número qualquer novamente') # exibe a mensagem de erro
```
