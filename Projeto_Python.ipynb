**Certifique-se de implementar um sistema de identificação único para cada funcionário.**

**Esse projeto permite aos alunos praticar e consolidar conceitos fundamentais em Python, como operadores, estruturas de controle, estruturas de dados e funções, enquanto constroem um aplicativo útil para gerenciar informações de funcionários de uma empresa fictícia.**


```python
f'''
Adicionar um novo funcionário ao sistema,
solicitando informações como nome, cargo, salário, e-mail, etc.

'''
#Adicionar um novo funcionário ao sistema, (Erica e Paulo)
#solicitando informações como nome, cargo, salário, e-mail, etc.


def incluir_funcionario():
    nome = input("Digite seu Nome: ")
    cargo = input("Digite seu Cargo: ")
    salario = float(input("Digite deu Salário: "))
    cidade = input("Digite sua Cidade: ")
    estado = input("Digite seu Estado: ")
    idade = int(input("Digite sua Idade: "))
    e_mail = input("Digite seu E-mail: ")

    funcionario = {
        "nome": nome,
        "cargo": cargo,
        "salario": salario,
        "cidade": cidade,
        "estado": estado,
        "idade": idade,
        "e_mail": e_mail,
    }

    funcionarios.append(funcionario) # Faz a inserção de novo funcionario ao final da lista

def consultar_funcionario(nome): # Função Consulta um unico funcionário por nome (Cleber)
    for funcionario in funcionarios:
        if funcionario["nome"] == nome:
            return funcionario
    return None

def listar_funcionarios(): # Função retorna uma lista com todos funconários (Eduardo)
    for funcionario in funcionarios:
        print(funcionario)

def deletar_funcionario(nome): # Função percorre a lista encontra o indice do funcionário e deleta, de acordo com nome solicita (Erivaldo)
    for i in range(len(funcionarios)):
        if funcionarios[i]["nome"] == nome:
            del funcionarios[i]
            return
    return None

funcionarios = [] # Cria a lista que receberá os funcionários

print('****** EMPRESA SALA_01 PROJETO PYTHON   ******\n')

while True: # Menu de escolha para chamar as funções (Tony)
    print("1. Incluir funcionário")
    print("2. Consultar funcionário")
    print("3. Listar funcionários")
    print("4. Deletar funcionário")
    print("0. Sair")

    escolha = int(input("Escolha uma opção: "))

    if escolha == 1:
        incluir_funcionario()
    elif escolha == 2:
        nome = input("Nome do funcionário: ")
        funcionario = consultar_funcionario(nome)
        if funcionario is not None:
            print(funcionario)
        else:
            print("Funcionário não encontrado.")
    elif escolha == 3:
        listar_funcionarios()
    elif escolha == 4:
        nome = input("Nome do funcionário: ")
        sucesso = deletar_funcionario(nome)
        if sucesso:
            print("Funcionário removido com sucesso.")
        else:
            print("Funcionário não encontrado.")
    elif escolha == 0:
        break
```
