## **Metodos para Strings**

```python
#sinal de + para concatenar as strings
nome = 'agnes'
sobrenome = 'fattobene'
print(nome + ' ' + sobrenome)
```

```python
#(len) o metodo que conta o numero de caractere de uma string
nome = 'agnes'
contagem =  len(nome)
print(nome, 'tem', contagem, 'caracteres')
```

```python
#(len) o metodo que conta o numero de caractere de uma string
nome = ' agnes '#espaço antes e depois do nome
contagem =  len(nome)
print(nome, 'tem', contagem, 'caracteres')#espaços tbm sao caracteres
```

```python
#(trip) metodo que elimina os caracteres indesejados de uma string
#por padrao ele remove espaços do inicio e do final de uma string
#print(le[nome])

nome = nome.strip()
```

```python
nome = nome.strip()
```

```python
print(len(nome))
```

```python
#escreva essa linha de codigo
preco = 'R$10,50'
print(preco)
```

```python
#(replace) metodo que substitui um caractere por outro
preco = preco.replace(',','.')
print(preco)
```

```python
nome = input("Informe seu nome: ")
print("O nome digitado é: ", nome)
```

```python
#(capitalize) Método para deixar a primeira letra da string maiscula
nome = nome.capitalize()
nome
```

```python
#metodo upper, metodo que coloca todos o caracteres em maiusculo
```

```python
# (upper) Método que coloca todas as letras em maiúsculo
nome = nome.upper()
nome
```

```python
# (casefold) Método para deixar tudo minusculo
nome = nome.casefold()
nome
```

```python
# (title) Método que coloca a primeira de letra de cada palavra e maiuscula
completo = 'agnes fattobene'
```

```python
completo = completo.title()
print(completo)
```

```python
#count) metodo para contar quantas vezes um caractere aparece em uma string
```

```python
# (count) Método para contar quantas vezes um caracter aparece em uma string
material = 'paralelepipedo'
contagem = material.count('p')
print('A quantidade de Ps é: ', contagem)
```

```python
# (endswith) Método que verifica se uma string termina com um caracter ou uma
# sequencia de caracteres
email = 'fran@gmail.com'
teste_fim = email.endswith('gmail.com')
print(teste_fim)
```
